2. Analyze Matrix A

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

A = np.array([[6, 5],[8, 6]])

print("Matrix A:")
print(A)
print("\nShape of A:", A.shape)
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)
rank_A = np.linalg.matrix_rank(A)
print("Rank of A:", rank_A)
eig_A = np.linalg.eigvals(A)
print("Eigenvalues of A:", eig_A)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:")
    print(inv_A)
```

4. Re-analyze Matrix B

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Matrix B
B = np.array([[7, 5],[8, 6]])

print("Matrix B:")
print(B)
print("\nShape of B:", B.shape)
det_B = np.linalg.det(B)
print("\nDeterminant of B:", det_B)
rank_B = np.linalg.matrix_rank(B)
print("Rank of B:", rank_B)
eig_B = np.linalg.eigvals(B)
print("Eigenvalues of B:", eig_B)
if det_B != 0:
    inv_B = np.linalg.inv(B)
    print("\nInverse of B:")
    print(inv_B)
```
